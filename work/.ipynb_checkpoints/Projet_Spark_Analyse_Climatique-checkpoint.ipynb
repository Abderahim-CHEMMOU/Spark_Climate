```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MonApp") \
    .config("spark.driver.memory", "12g") \
    .config("spark.executor.memory", "12g") \
    .config("spark.driver.maxResultSize", "12g") \
    .master("local[*]") \
    .getOrCreate()
```

## Heure 3 : Transformation des données


```python
from pyspark.sql import functions as F
df_transformed = df_cleaned.withColumn('year', F.year(F.col('DATE'))) \
                           .withColumn('month', F.month(F.col('DATE'))) \
                           .withColumn('day', F.dayofmonth(F.col('DATE')))

# Classifier les jours chauds
df_transformed = df_transformed.withColumn('is_hot', F.col('TEMP') > 30)
```

```python
# Afficher les 5 premiers enregistrements
# 1. Quels sont les 5 premiers enregistrements après le parsing de la colonne `date` ?
df_transformed.select('DATE', 'year', 'month', 'day', 'TEMP', 'is_hot').show(5)
```

```python
# Compter les jours chauds
hot_days_count = df_transformed.filter(F.col('is_hot') == True).count()
print(f"Nombre de jours chauds : {hot_days_count}")
```

## Heure 4 : Analyse exploratoire des données (EDA)


```python
# Calculer les moyennes annuelles
# Filtrer les données pour l'année 2020
avg_temp_2020 = df_transformed.filter(F.col('year') == 2020).agg(F.avg('TEMP').alias('avg_temp_2020'))
avg_temp_2020.show()
```

```python
# Calculer la température moyenne par station
avg_temp_by_station = df_transformed.groupBy('STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION') \
    .agg(F.avg('TEMP').alias('avg_temp')) \
    .orderBy(F.desc('avg_temp'))

# Obtenir les 5 stations avec les températures moyennes les plus élevées
top_5_stations = avg_temp_by_station.limit(5)
#températures moyennes les plus élevées
top_5_stations.show(truncate=False)
```

```python
# Calculer la moyenne annuelle des précipitations
precipitation_by_year = df_transformed.groupBy('year').avg('PRCP')
precipitation_by_year = precipitation_by_year.orderBy('year')
precipitation_by_year.show(truncate=False)
```

```python
import matplotlib.pyplot as plt
precipitation_pd = precipitation_by_year.toPandas()
plt.figure(figsize=(12, 6))
plt.plot(precipitation_pd['year'], precipitation_pd['avg(PRCP)'], marker='o')
plt.title('Évolution des précipitations mondiales (1970-2020)')
plt.xlabel('Année')
plt.ylabel('Précipitations moyennes (mm)')
plt.grid(True)
plt.show()
```

## Heure 5 : Requêtes avancées


```python
# Enregistrer comme vue SQL temporaire et exécuter des requêtes
# df_transformed.createOrReplaceTempView('climate')
# spark.sql('SELECT ...')
```

```python
df_transformed.createOrReplaceTempView('climate')

# Année la plus froide
coldest_year = spark.sql("""
    SELECT year, AVG(TEMP) as avg_temp
    FROM climate
    GROUP BY year
    ORDER BY avg_temp ASC
    LIMIT 1
""")
coldest_year.show()
```

```python
# Station avec le plus grand nombre d'enregistrements
busiest_station = spark.sql("""
    SELECT STATION, NAME, COUNT(*) as record_count
    FROM climate
    GROUP BY STATION, NAME
    ORDER BY record_count DESC
    LIMIT 1
""")
busiest_station.show()
```

## Heure 6 : Visualisation


```python
# Exporter les données vers Pandas et créer des visualisations
# import matplotlib.pyplot as plt
# df_pandas = avg_temp_by_year.toPandas()
# plt.plot(df_pandas['year'], df_pandas['avg_temperature'])
```

```python
import matplotlib.pyplot as plt

# Calculer la température moyenne par année
avg_temp_by_year = df_transformed.groupBy('year').agg(F.avg('TEMP').alias('avg_temperature')).orderBy('year')
df_pandas = avg_temp_by_year.toPandas()

# Graphique de la température moyenne par année
plt.figure(figsize=(10, 6))
plt.plot(df_pandas['year'], df_pandas['avg_temperature'])
plt.xlabel('Année')
plt.ylabel('Température moyenne')
plt.title('Évolution de la température moyenne mondiale')
plt.show()
```

#### La hausse est particulièrement marquée après 2022, ce qui pourrait indiquer une accélération du réchauffement climatique.

```python
# Calculer les précipitations moyennes par année
avg_prcp_by_year = df_transformed.groupBy('year').agg(F.avg('PRCP').alias('avg_prcp')).orderBy('year')
prcp_pandas = avg_prcp_by_year.toPandas()

# Histogramme des précipitations
plt.figure(figsize=(10, 6))
plt.bar(prcp_pandas['year'], prcp_pandas['avg_prcp'], color='blue')
plt.xlabel('Année')
plt.ylabel('Précipitations moyennes (mm)')
plt.title('Évolution des précipitations mondiales annuelles')
plt.show()
```

```python
# Grouper par latitude (approximativement par zone climatique)
df_transformed = df_transformed.withColumn('latitude_zone', 
                                           F.when(F.col('LATITUDE') > 30, 'Nord')
                                            .when(F.col('LATITUDE') < -30, 'Sud')
                                            .otherwise('Équateur'))

# Calculer les températures moyennes par latitude et année
avg_temp_by_latitude = df_transformed.groupBy('year', 'latitude_zone') \
                                     .agg(F.avg('TEMP').alias('avg_temperature')) \
                                     .orderBy('year', 'latitude_zone')

avg_temp_by_latitude.show()
```

```python
# Convertir en Pandas pour visualisation
latitude_pandas = avg_temp_by_latitude.toPandas()

# Tracer les températures par zones climatiques
plt.figure(figsize=(12, 8))
for zone in latitude_pandas['latitude_zone'].unique():
    subset = latitude_pandas[latitude_pandas['latitude_zone'] == zone]
    plt.plot(subset['year'], subset['avg_temperature'], label=f'Zone {zone}')
    
plt.xlabel('Année')
plt.ylabel('Température moyenne (°F)')
plt.title('Évolution des températures par zones climatiques')
plt.legend()
plt.grid(True)
plt.show()
```

#### Calculer la médiane pour chaque année 

```python
# Calculer la médiane pour chaque année
median_temp_by_year = df_transformed.groupBy('year').agg(F.expr('percentile_approx(TEMP, 0.5)').alias('median_temp'))
median_temp_by_year.show()
```
